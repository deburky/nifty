```python
adj = joblib.load('adj.pkl')
```

```python
G = nx.Graph()

for i in range(adj.shape[0]):
    G.add_node(i)

non_zero_elements = adj.nonzero()

for i, j in zip(non_zero_elements[0], non_zero_elements[1]):
    G.add_edge(i, j, weight=adj[i, j])
```

```python
nx.draw(G, with_labels=False, node_size=50, linewidths=0.5)
```
